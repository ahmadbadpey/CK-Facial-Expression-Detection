```python
from mtcnn import MTCNN
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import img_to_array
import numpy as np
import cv2 as cv
```

```python
model = load_model('ck_plus_model.h5')
states = ['anger','contempt','disgust','fear','happy','sadness','surprise']
```

```python
def face_detect(img):
    rgb_img = cv.cvtColor(img , cv.COLOR_BGR2RGB)

    face = detector_obj.detect_faces(rgb_img)[0]
    x , y , w , h = face['box']

    return rgb_img[y:y+h , x:x+w] , x , y , w , h # This crop only section of image that contain person Face

def preprocess(face):
    norm_face = cv.resize(face , (48,48))
    norm_face = norm_face/255.0
    return np.array([norm_face])
```

```python
video = cv.VideoCapture(0)
detector_obj= MTCNN()

while True:
    success , img = video.read() #success value to make sure it's reading the frames successfully.

    face , x , y , w , h= face_detect(img)
    normalized_face = preprocess(face)
    out = model.predict(normalized_face)[0]
    max_index = np.argmax(out)
    text = states[max_index]
    
    cv.rectangle(img , (x, y ) , (x+ w , y+ h) , (0,255,0) , 2)
    cv.putText(img , f"State : {text}" , (x , y - 10)  , cv.FONT_HERSHEY_PLAIN,  1 , (255,255,0) , 1 )
    
    cv.imshow('MyWin',img)
    if cv.waitKey(30) == ord('q'):
        break

cv.destroyAllWindows()
```
